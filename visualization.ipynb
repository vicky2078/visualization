```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('iris.csv')
```

```python
data.head()
```

# **1.Scatterplot:**

```python
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.title('Scatterplot for bivariant analysis')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

```python
plt.scatter(data['petal_length'],data['petal_width'],color='red')
plt.title('Scatterplot for bivariant analysis')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

# **2.Histogram:**

```python
plt.hist(data['sepal_length'],bins=20)
plt.title('Histogram for univariate analysis')
plt.xlabel('sepal_length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data['sepal_length'])
plt.title('Histogram for univariate analysis')
plt.xlabel('sepal_length')
plt.ylabel('Frequency')
plt.show()
```

# **3.Bar Chart:**

```python
sns.barplot(data['species'])
plt.title('Bar Chart for categorical data')
```

# **4.Pie Chart:**

```python
lables=['setosa','versicolor','virginica']
sizes=[50,50,50]
plt.pie(sizes,labels=lables,explode=(0.1,0.1,0.1))
plt.title('Pie Chart for categorical data')
plt.show()
```

# **5.Countplot:**

```python
sns.countplot(data['species'])
plt.title('Countplot for categorical data')
```

# **6.Boxplot:**

```python
sns.boxplot(data['sepal_length'])
plt.title('Boxplot for univariate analysis')
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_length',data=data)
plt.title('Boxplot for bivariant analysis')
```

# **7.Heatmap:**

```python
data1=data.drop('species',axis=1)
data1.corr()
```

```python
sns.heatmap(data1.corr(),cmap='coolwarm',annot=True)
plt.title('Heatmap for bivariant analysis')
```

# **8.Distplot:**

```python
sns.displot(data['sepal_length'],bins=25,kde=True)
plt.title('Distplot for univariate analysis')
```

```python
sns.displot(data['sepal_width'],bins=25,kde=True)
plt.title('Distplot for univariate analysis')
```

# **9.Jointplot:**

```python
sns.jointplot(x='sepal_length',y='petal_length',data=data,kind='scatter')
```
